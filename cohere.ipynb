```python
# Step 1: Install Cohere
!pip install cohere

# Step 2: Import Libraries and Initialize the Client
import cohere
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Replace 'your-api-key' with your actual Cohere API key
api_key = 'zQVSgu4XjSjNmCBTm1xLwQz9CwJXP83PNVjoMfAE'
co = cohere.Client(api_key)

# Step 3: Define Document Corpus with Sources
documents = [
    {"text": "The Eiffel Tower is located in Paris, France. It was completed in 1889 and is a global cultural icon of France.", "source": "Eiffel Tower Facts"},
    {"text": "The Great Wall of China is a series of fortifications made of stone, brick, tamped earth, wood, and other materials, built along the northern borders of China to protect against various nomadic groups.", "source": "Great Wall of China History"},
    {"text": "The Pyramids of Giza are ancient pyramid structures located in Giza, Egypt. They were built during the Fourth Dynasty of the Old Kingdom of Ancient Egypt.", "source": "Pyramids of Giza Overview"}
]

# Step 4: Embed Documents
def embed_documents(documents):
    texts = [doc['text'] for doc in documents]
    response = co.embed(texts=texts, model='embed-english-v2.0')
    return response.embeddings

document_embeddings = embed_documents(documents)

# Step 5: Implement Document Retrieval with Source Citation
def retrieve_documents(query, document_embeddings, documents, top_k=1):
    query_embedding = co.embed(texts=[query], model='embed-english-v2.0').embeddings
    similarities = cosine_similarity(query_embedding, document_embeddings)[0]
    top_indices = np.argsort(similarities)[::-1][:top_k]
    return [(documents[i]['text'], documents[i]['source']) for i in top_indices]

# Step 6: Generate Responses Using Retrieved Information and Cite Sources
def generate_response(query, documents):
    retrieved_docs_with_sources = retrieve_documents(query, document_embeddings, documents, top_k=1)
    retrieved_docs = [doc[0] for doc in retrieved_docs_with_sources]
    sources = [doc[1] for doc in retrieved_docs_with_sources]
    context = " ".join(retrieved_docs)
    prompt = f"Context: {context}\n\nQuestion: {query}\n\nAnswer:"

    response = co.generate(
        model='command-xlarge-nightly',
        prompt=prompt,
        max_tokens=100,
        temperature=0.7,
        stop_sequences=["\n"]
    )
    cited_response = response.generations[0].text.strip()
    cited_response += f"\n\nSources: {', '.join(sources)}"
    return cited_response

# Example usage
query = "Where is the Eiffel Tower located?"
response = generate_response(query, documents)
print("Generated Response: ", response)
```
